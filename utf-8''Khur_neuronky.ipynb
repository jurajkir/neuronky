```python
#1. import kniznic na spracovanie dat a suborov
import numpy as np 
import pandas as pd
```

```python
#2. ocistenie dat
df = pd.read_csv('tesla_april.csv', parse_dates=['datetime'], infer_datetime_format=True) #upload a nacitanie prveho csv suboru pod nazvom df
```

```python
df['date'] = df.datetime.dt.date #vytvorenie stlpca date- jeho naplnenie zo stlpca datetime ale iba datumom
df['time'] = df.datetime.dt.time #vytvorenie stlpca time- jeho naplnenie zo stlpca datetime ale iba casom
df #vypis csv
```

```python
df['date'] = pd.to_datetime(df['date']) #konvertovanie stlpcu date na typ datetime
```

```python
df1=(df[df["date"]=='2020-03-30'][:160]) #deklarovanie novej premennej df1- k nej bolo priradenych prvych 160 riadkov zo stlpca date z premennej df, ktore sa rovnali datumu 2020-03-30
df1 #vypis premennej df1
```

```python
df2=(df[df["date"]=='2020-04-29'][:160]) #deklarovanie novej premennej df2- k nej bolo priradenych prvych 160 riadkov zo stlpca date z premennej df, ktore sa rovnali datumu 2020-04-29
df2 #vypis premennej df2
```

```python
may = pd.read_csv('tesla_may.csv', parse_dates=['datetime'], infer_datetime_format=True) #nacitanie noveho csv suboru do premennej may
```

```python
may['date'] = may.datetime.dt.date #vytvorenie stlpca date- jeho naplnenie zo stlpca datetime ale iba datumom
may['time'] = may.datetime.dt.time #vytvorenie stlpca time- jeho naplnenie zo stlpca datetime ale iba casom
may['date'] = pd.to_datetime(may['date']) #konvertovanie stlpcu date na typ datetime
```

```python
df3=(may[may["date"]=='2020-05-01'][:160]) #deklarovanie novej premennej df3- k nej bolo priradenych prvych 160 riadkov zo stlpca date z premennej may, ktore sa rovnali datumu 2020-05-01
df3 #vypis premennej df1
```

```python
ocistene =df1[['open','high','low','close','date']] #vytvorenie premennej ocistene- jej naplnenie premennou df1 ale iba stlpcami open, high, low, close a date
ocistene #vypisanie premennej
```

```python
ocistene1 =df2[['open','high','low','close','date']] #vytvorenie premennej ocistene1- jej naplnenie premennou df2 ale iba stlpcami open, high, low, close a date
ocistene1 #vypis premennej
```

```python
ocistene2 =df3[['open','high','low','close','date']] #vytvorenie premennej ocistene2- jej naplnenie premennou df3 ale iba stlpcami open, high, low, close a date
ocistene2 #vypis premennej
```

```python
df_final=pd.concat((ocistene,ocistene1,ocistene2)) #deklaracia premennej df_final- jej naplnenie predoslymi troma premennymi
df_final #vypis obsahu premennej df_final
```

```python
#3. importovanie kniznic na pracu v kerasi
from keras.models import Sequential 
from keras.layers import Dense
from keras.utils import np_utils
from keras.utils import to_categorical
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
#4. rozdelenie na vstupne (X) a vystupne (y) premenne
X = df_final.iloc[:,[0,3]].values #premennej X priradi vsetky hodnoty zo stlpca s indexom 0 a 3
print(df_final.shape)
y = [-1 if row[1] - row[0] < 0 else 1 for row in X] #podmienka na pridelenie 1/-1 premennej y
y = y[1:] #posunutie dni o jeden
y.append(0) #poslednemu dnu priradi 0

df_final[5] = y #vytvori piaty stlpec kde priradi hodnoty premennej y na zaklade predoslej podmienky
```

```python
train,test = train_test_split(df_final, shuffle=True, train_size=0.8, test_size=0.2) #rozdelenie trenovacich a testovacich dat v pomere 80/20%.
print(train.shape,test.shape)
```

```python
#5. vytvorenie modelu neuronovej siete
model = Sequential() #vytvorenie sekvenčneho modelu a pridanie vrstiev v poradí výpočtu
model.add(Dense(128,input_dim=4, activation='relu')) #input_dim- pocet vstupnych atributov
model.add(Dense(256,input_dim=128, activation='relu')) 
model.add(Dense(64, input_dim=256, activation='sigmoid'))
model.add(Dense(1))

#6. kompilacia modelu neuronovej siete
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) 
```

```python
train_x = train.iloc[:,0:4].values #priradi premennej train_x vsetky data zo stlpca 0 az 3 vratane
train_y = train.iloc[:, 5].values #priradi premennej train_y data z 5 stlpca

print(train_x.shape, train_y.shape) #vypis trenovani

history = model.fit(train_x, train_y, epochs=100, batch_size=10, verbose=1) #vytvorenie grafu na prezentaciu priebehu trenovania
```

```python
print(history.history) #vypisanie historie ucenia chybovosti a presnosti
```

```python
test_x, test_y = test.iloc[:,0:4].values, test.iloc[:,5].values #priradenie tych istych dat premennym test_x/y ako maju premenne train_x/y
loss, accuracy = model.evaluate(test_x, test_y) 
```

```python
test_x, test_y = test.iloc[:,0:4].values, test.iloc[:,5].values
model.evaluate(test_x, test_y) #vypis chybovosti 
```

```python
print('Presnost je:', accuracy*100) #vypis presnosti
```

```python
#graf na zhrnutie historie presnosti ucenia
plt.plot(history.history['accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
#graf na zhrnutie historie presnosti chybovosti
plt.plot(history.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
